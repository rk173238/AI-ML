```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel('C:/Users/RO361326/scikit_learn_data/mldata/data.xlsx')
df.columns = [['mpg','cylinder','disp','hp','weight','acc','year','origin','name']]
```

```python
df.head()
```

# univariate regression

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = df['acc']
y = df['mpg']
x_train,x_test,y_train,y_test = train_test_split(X,y,train_size = 0.8,random_state = 42)
```

```python
x_tr = (np.array(x_train)).reshape(len(x_train),1)
x_ts = (np.array(x_test)).reshape(len(x_test),1)
y_tr = (np.array(y_train)).reshape(len(y_train),1)
y_ts = (np.array(y_test)).reshape(len(y_test),1)
```

```python
plt.scatter(x_ts,y_ts)
```

```python
plt.plot(x_tr,y_tr)
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_tr,y_tr)
print(reg.score(x_tr,y_tr))
print(reg.coef_,reg.intercept_)
```

```python
p = reg.predict(x_train)
plt.scatter(x_ts,y_ts)
```

```python
plt.scatter(x_tr,y_tr)
plt.plot(x_tr,p,color = 'g')
```

```python
p = reg.predict(x_ts)
plt.plot(x_ts,p,color = 'g')
plt.scatter(x_ts,y_ts)
reg.score(x_ts,y_ts)
```
